```python
import numpy as np
import pandas as pd

df = pd.read_excel('Data Model - Pizza Sales.xlsx')
df.head()
```

```python
# Check for missing data in each column
missing_columns = df.columns[df.isnull().any()].tolist()
print("Columns with missing data:", missing_columns)
```

```python
df.columns
```

```python
# Identify columns with corrupted data
corrupted_columns = []
numeric_columns = df.select_dtypes(include=['float', 'int']).columns.tolist()

for column in numeric_columns:
    try:
        df[column].astype(float)
    except ValueError:
        corrupted_columns.append(column)
print("Columns with potentially corrupted data:", corrupted_columns)
```

```python
#check for corrupted data in columns with string data type
corrupted_string_columns = []
for column in df.select_dtypes(include='object').columns:
    try:
        df[column].astype(str)
    except UnicodeDecodeError:
        corrupted_string_columns.append(column)
print("Columns with potentially corrupted string data:", corrupted_string_columns)
```

```python
df['quantity'].fillna(df['quantity'].mean(), inplace=True)
#to replace any NaN values with mean of column- although there dont seem to be any
```
